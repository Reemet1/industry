```python
#pandas.DataFrame( data, index, columns, dtype, copy)

#Creating DataFrames from Lists, dict, Series, Numpy ndarrays, another DataFrame

#Creating empty DataFrame
df = pd.DataFrame()

#Creating df from a List
pd.DataFrame([1,2,3,4,5])

data = [['Alex',10],['Bob',12],['Clarke',13]]
pd.DataFrame(data,columns=['Name','Age'], dtype=float)

#Creating df from dict of ndarrays or lists
data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
pd.DataFrame(data)

#Creating indexed df using arrays
data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
pd.DataFrame(data, index=['rank1','rank2','rank3','rank4'])

#Creating df from List of Dicts
data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]
pd.DataFrame(data)

#Same as previous but with indexes
data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]
pd.DataFrame(data, index=['first', 'second'])

#Same as previous but also with columns
data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]
pd.DataFrame(data, index=['first', 'second'], columns=['a','b','c'])

#Creating df as Dict of Series
d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']),
   'two' : pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}
pd.DataFrame(d)
```

```python
df = pd.DataFrame(np.random.randn(5, 3), index=['a', 'c', 'e', 'f',
'h'],columns=['one', 'two', 'three'])

df = df.reindex(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])

#Checking for missing values
df['one'].isnull()
df['one'].notnull()

#When summing data, NA will be treated as Zero. If the data are all NA, then the result will be NA.
df['one'].sum()

#Replace NaN with a Scalar Value
df.fillna(0)

#Drop columns with missing values
df.dropna()

#Drop rows with missing values
df.dropna(axis=1)

#Replace Missing (or) Generic Values
df = pd.DataFrame({'one':[10,20,30,40,50,2000], 'two':[1000,0,30,40,50,60]})
df.replace({1000:10,2000:60})
```
