```python
#Loo funktsioon, mis võtab sisendiks inimese kaalu ja pikkus ning kuvab vastavalt kehamassiindeksile, millisesse kaalukategooriasse inimene (ülevaloleva tabeli järgi) kuulub
#Kehamassi indeksi arvutamise jaoks saad kasutada juba varasemalt loodud funktsiooni, s.t sa saad kutsuda välja ühte funktsiooni enda loodava funktsiooni sees.

def kehamassi_index(kaal_kg, pikkus_m):
  tulemus = kaal_kg/(pikkus_m**2)
  return tulemus

def kaalukategooria(kaal, pikkus):
  kmi = kehamassi_index(kaal, pikkus)
  if kmi < 16:
    print("Tõsine alakaal")
  elif (16 <= kmi and kmi < 17):
    print("Keskmine alakaal")
  elif (17 <= kmi and kmi < 18.5):
    print("Kerge alakaal")
  elif (18.5 <= kmi and kmi < 25):
    print("Normaalkaal")
  elif (25 <= kmi and kmi < 30):
    print("Ülekaal")
  elif (30 <= kmi and kmi < 35):
    print("Rasvumise 1. klass")
  elif (35 <= kmi and kmi < 40):
    print("Rasvumise 2. klass")
  else:
    print("Rasvumise 3. klass")
 
kaalukategooria(90, 1.8)
```
