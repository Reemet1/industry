## Järjendid

Sageli ei tule meil tegemist teha üksikute arvude ja sõnedega, vaid arvude ja sõnede kogumitega.

```python
isik1 = "Andrus"
isik2 = "Kalmer"
isik3 = "Maris"
isik4 = "Kersti"
isik5 = "Taavi"

print(isik1)
print(isik2)
print(isik3)
print(isik4)
print(isik5)

isikud = ["Andrus", "Kalmer", "Maris", "Kersti", "Taavi"]

print(isikud)

print(isikud[0])

isikud.append("Kermo")

print(isikud)

for isik in isikud:
  print(isik)
```

```python
#Järjendite puhul on võimalik järjest läbi käia kõik selle järjendi elemendid (liikmed) ning iga elemendiga midagi teha
for isik in isikud:
  print(isik)

arvud = [1,2,3,4,5]
for arv in arvud:
  print("Arvu", arv, "ruut on", arv*arv)
```

## Tingimuslaused

Tingimuslaused programmi töö huvitavaks. Tingimuslausete abil ei toimu programmi töö enam ühte teed pidi, vaid võib käituda erinevalt vastavalt tingimustele. 

```python
vanus = 22

if (vanus < 18):
  print("Alaealine")
else:
  print("Täisleaine")
```

## Tsüklid

Tsüklid võimaldavad mingit koodijuppi jooksutada mitu korda. Oletame, et meil on vaja midagi teha järjendi iga elemendiga. Siis saab tsüklite abil mugavalt järjendi iga elemendi läbi käia ning iga elemendi puhul midagi teha.

```python
i = 1
while (i < 10):
  print(i)
  i = i+1 #Suurendame i väärtust ühe võrra. Vastasel juhul jääks tsükkel lõpmatult ketrama.
```

```python
#Arvutame esimese 100 täisarvu summa

i = 1
summa = 0
while (i <= 100):
  summa += i
  i = i+1

print("Esimese 100 täisarvu summa on", summa)
```

```python
arvud = [1,2,3,4,5]

for arv in arvud:
  print("Arvu", str(arv), "ruut on " + str(arv**2))
```

## Funktsioonid

Funktsioonid võimaldavad eraldada mingit funktsionaalsust ja koodi, ning sellele koodile hiljem programmi töö käigus nimeliselt viidata. 

```python
#Oletame, et me tahame tervitada mitut inimest. 
print("Tere, Kalmer")
print("Tere, Salme")
print("Tere, Juku")
print("Tere, Mari")
print("Tere, Joosep")

#Kui me kordamine samat koodijuppi mitu korda, siis on võimalik see koodijupp viia eraldi kohta nimega funktsioon

def tervita(nimi):
  print("Tere", nimi)

tervita("Reemet")
```

```python
def kehamassi_index(kaal_kg, pikkus_m):
  tulemus = kaal_kg/(pikkus_m**2)
  return tulemus

bmi = kehamassi_index(90, 1.80)

print("Sinu kehamassi indeks on", bmi)

#Loo uus funktsiooni kaaluhinnang(...), mis võtan argumendiks kehamassi indeksi ning tagastab sõnalise hinnangu kehakaalule sõltuvalt, millisesse vahemikku kehamassi indeks jääb
```

##Funktsiooni argumendid
Me ei taha luua iga tervituse jaoks eraldi funktsiooni, vaid tahame, et oleks vaid üks funktsioon, mille abil oleks võimalik tervitada ükskõik keda. Selle saavutamiseks on funktsioonile lisada argumente, mille väärtuseid funktsiooni sees saab dünaamiliselt kasutada ning seejärel ühte ja samat funktsiooni erinevate argumentide väärtustega välja kutsuda.

```python
def tervita(nimi):
  print("Tere", nimi)

tervita("Reemet")
```

##Funktsioonid, mis tagastavad väärtuseid

```python
def arvutaKeskmine(a, b):
  keskmine = (a+b)/2
  return keskmine

kesk = arvutaKeskmine(4, 6)
print("Funktsiooni poolt tagastatud väärtus on", kesk)
```
