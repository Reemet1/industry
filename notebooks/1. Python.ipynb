###If-else lause
If-else lauset kasutatakse juhul, kui if-osas olev tingimus ei kehti (ja seda koodiosa ei täideta), kuid soovitakse, et sellisel juhul täidetakse mingit teist (else osas olevat) koodi.

````python
vanus = 22

if (vanus >= 18):
  print("Täisearline")
else:
  print("Alaealine")
````

###If-elseif-else lause
Kui me tahame kontrollida kuitahes mitu tõeväärtust, siis saame need lisada järjest "elif (tõeväärtus)" juppidena.

````python
kuu = 5

if (kuu == 1):
  print("Jaanuar")
elif (kuu == 2):
  print("Veebruar")
elif (kuu == 3):
  print("Märts")
elif (kuu == 4):
  print("Aprill")
elif (kuu == 5):
  print("Mai")
elif (kuu == 6):
  print("Juuni")
elif (kuu == 7):
  print("Juuli")
elif (kuu == 8):
  print("August")
elif (kuu == 9):
  print("September")
elif (kuu == 10):
  print("Oktoober")
elif (kuu == 11):
  print("November")
elif (kuu == 12):
  print("Detsember")
else:
  print("Sisestatud kuu number ei ole sobiv. Arv peab olema 1-st 12-ni")
````

##Funktsioonid, mis tagastavad väärtusi
Sageli on vaja funktsiooni töö tulemusi ja arvutusi mujal koodis edasi kasutada. Selle saavutamiseks on võimalik funktsiooni lõppemisel sealt tagastada vajaminevaid väärtuseid, mida saab eraldi muutujasse salvestada. Funktsioonist saab väärtuseid kasutada märksõnaga 'return', mille järele tuleb lisada väärtuse, mida soovid tagastada, nt 'return 10'.

````python
def arvutaKeskmine(a, b):
  keskmine = (a+b)/2 
  return keskmine #Tagastame funktsiooni töö lõpus muutuja 'keskmine' väärtuse

kesk = arvutaKeskmine(4, 6) #Funktsioonist tagastatud muutuja 'keskmine' väärtuse saab nüüd salvestada endale sobivasse muutujasse ja edaspidi seda kasutada.
print("Funktsiooni poolt tagastatud väärtus on", kesk) #Kasutame funktsioon "arvutaKeskmine" poolt tagastatud ja salvestatud väärtust.
````

````python
def kaalukategooria(kaal, pikkus):
  kmi = kehamassi_index(kaal, pikkus)
  if kmi < 16:
    print("Tõsine alakaal")
  elif (16 <= kmi and kmi < 17):
    print("Keskmine alakaal")
  elif (17 <= kmi and kmi < 18.5):
    print("Kerge alakaal")
  elif (18.5 <= kmi and kmi < 25):
    print("Normaalkaal")
  elif (25 <= kmi and kmi < 30):
    print("Ülekaal")
  elif (30 <= kmi and kmi < 35):
    print("Rasvumise 1. klass")
  elif (35 <= kmi and kmi < 40):
    print("Rasvumise 2. klass")
  else:
    print("Rasvumise 3. klass")
 
kaalukategooria(90, 1.8)
````
